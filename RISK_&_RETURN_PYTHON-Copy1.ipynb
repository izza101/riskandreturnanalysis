```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import yfinance as yt
from pandas_datareader import data as wb

plt.rcParams['figure.figsize'] = 12,8
```

```python
ICBP
```

### Task 3 : Visualization 1 Stock (Close)

```python
ICBP['Close'].plot(figsize = (12,8))
plt.title("Pergerakan Harga Saham ICBP (Daily)", fontsize = 18)
plt.ylabel("Harga Saham")
plt.show()
```

### Task 4 : Daily Return Using equation

return = (P1-P0) / P0

```python
ICBP_daily_r = ((ICBP['Close']-ICBP['Close'].shift(1)) / ICBP['Close'].shift(1)).mul(100)
```

```python
ICBP_daily_r.plot(figsize = (12,8))
plt.show()
```

### Task 5 : Daily Return Using Method

pct_change

```python
ICBP_daily_r = ICBP['Close'].pct_change(1).mul(100)
```

```python
ICBP_daily_r.plot(figsize = (12,8))
plt.show()
```

### Task 6 : Daily Return (Mean)

```python
ICBP_daily_r_mean = ICBP_daily_r.mean()
```

```python
ICBP_daily_r_mean
```

### Task 7 : Annualize Daily Return

```python
ICBP_daily_r_mean*252
```

### Task 8 : Visualitation Frequency Daily Return 

```python
ICBP_daily_r.plot(kind = 'hist', figsize = (12,8), bins = 100)
plt.show()
```

### Task 9 : Count Standart Deviation Using std (Daily) 

```python
ICBP_std_daily = ICBP_daily_r.std()
```

```python
ICBP_std_daily
```

### Task 10 : Count Standart Devuation Using std (Annualize)

```python
ICBP_std_ann = ICBP_daily_r.std()*np.sqrt(252)
```

```python
ICBP_std_ann
```

### Task 12 : Scatter Plot ICBP vs IHSG

```python
data = wb.DataReader(['^JKSE','ICBP.JK'], data_source = 'yahoo', start = '2014-01-01')['Close']
```

```python
data_return = data.pct_change(1).mul(100).dropna()
```

```python
summary = data_return.describe()
```

```python
summary = summary.T.loc[:,['mean','std']]
```

```python
summary
```

```python
summary["ann.mean"] = summary['mean']*252
summary["ann.std"] = summary['std']*np.sqrt(252)
```

```python
summary.drop(columns = ["mean", "std"], inplace = True)
```

```python
summary
```

```python
summary.plot.scatter(x = "ann.std", y = "ann.mean", figsize = (12,8), s = 100, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy = (summary.loc[i,"ann.std"]+0.002, summary.loc[i, "ann.mean"]+0.002), size = 15)
plt.xlabel("ann.Risk(std)", fontsize = 15)
plt.ylabel("ann.Return(mean)", fontsize = 15)
plt.title("Risk/Return IHSG vs ICBP", fontsize = 20)
plt.show()
```
